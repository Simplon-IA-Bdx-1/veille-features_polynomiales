# Regression Polynomiale avec la librairie Scikit-learn

### Création d'un jeu de données

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

### Visualisation des données

```python
n_samples = 100
X = np.linspace(0, 10, 100)
y = X ** 3 + np.random.randn(n_samples) * 100 + 100

plt.figure(figsize=(10,8))
plt.scatter(X, y)
```

### Regression linéaire

```python
lin_reg = LinearRegression()
lin_reg.fit(X.reshape(-1, 1), y)

model_pred = lin_reg.predict(X.reshape(-1,1))

plt.figure(figsize=(10,8));
plt.scatter(X, y);
plt.plot(X, model_pred);
print(r2_score(y, model_pred))
```

### Régression polynomiale de degrés 2

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X.reshape(-1, 1))
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y.reshape(-1, 1))

y_pred = lin_reg_2.predict(X_poly)

plt.figure(figsize=(10,8));
plt.scatter(X, y);
plt.plot(X, y_pred);
print(r2_score(y, y_pred))
```

### Régression polynomiale de degrés 3

```python
poly_reg_2 = PolynomialFeatures(degree=3)
X_poly_2 = poly_reg_2.fit_transform(X.reshape(-1, 1))
```

```python
lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly_2, y.reshape(-1, 1))

y_pred = lin_reg_3.predict(X_poly_2)

plt.figure(figsize=(10,8));
plt.scatter(X, y);
plt.plot(X, y_pred);
print(r2_score(y, y_pred))
```
